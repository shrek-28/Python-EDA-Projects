### *Import of Libraries*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### *Import of Dataset*

```python
df = pd.read_csv("./Top Youtubers Dataset.csv", encoding = "ISO-8859-1")
```

```python
df.head()
```

### *Data Columns Details*

Rank: Details the rank of YouTube channel in terms of subscribers

YouTuber: Channel Name

Subscribers: Tells the number of subscriberes of a YouTube Channel

Video Views: The number of views every YouTuber has

Video Count: The number of videos uploaded by a YouTuber

Category: The type of content a channel posts

Started: The year that the channel has started

#### Checking the Effects Each Numeric Value has over the other

```python
df.corr(numeric_only=True)
```

```python
data = df.corr(numeric_only=True)
sns.heatmap(data=data, annot=True)
```

#### Variations in Views by Top 10 YouTubers

```python
top_10_views = df.sort_values(by='Video Views', ascending=False).head(10)
sns.barplot(data=top_10_views,
            y='Youtuber',
            x='Video Views');
plt.title('Top 10 YouTube Views by YouTuber');
```

#### Top 10 YouTubers by Subscribers

```python
top_10_subscribers = df.sort_values(by='Subscribers', ascending=False).head(10)
sns.barplot(data=top_10_subscribers,
            y='Youtuber',
            x='Video Views');
plt.title('Top 10 YouTubers - by Subscribers');
```

#### Top 10 YouTube Channels by Video Count

```python
top_10_vid_count = df.sort_values(by='Video Count', ascending=False).head(10)
sns.barplot(data=top_10_vid_count,
            y='Youtuber',
            x='Video Count');
plt.title('Top 10 YouTubers - by Video Count');
```

#### YouTubers with the Lowest View Count among the Top 1000

```python
low_10_views = df.sort_values(by='Video Views', ascending=True).head(10)
sns.barplot(data=low_10_views,
            y='Youtuber',
            x='Video Views');
plt.title('YouTubers with the Lowest View count in the ');
```

#### YouTubers with the Lowest Subscriber Count in the Top 1000

```python
low_10_vid_count = df.sort_values(by='Video Count', ascending=True).head(20)
sns.barplot(data=low_10_vid_count,
            y='Youtuber',
            x='Video Count');
plt.title('YouTubers with the Lowest Video Count in the Top 1000');
```
